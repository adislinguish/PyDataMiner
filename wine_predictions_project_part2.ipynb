<h2>Summary Description</h2>
This dataset aggregates detailed information on wine products and user ratings sourced from the open web. It includes ratings linked to wines from various countries, regions, and wineries spanning a 5-year period (2017–2021). Each wine entry is described by attributes such as type, grape varieties, alcohol content, body, acidity, and food pairings. The dataset also includes metadata like vintage years, country of origin, and winery websites. The ratings file connects users to specific wines and vintages, providing a rich foundation for analyzing consumer preferences and wine characteristics.

<h3>Data Type Conversion</h3>
How the dataset does not have lot of numeric values, we are going to convert the ranked categorical variables of Acidity and Body to numerical ranking. 
Converting Acidity to numerical ranked {Low = 1, Medium = 2, High = 3}

```python
acidityConvertion = lambda x: (
    1 if x.lower() == 'low' else
    2 if x.lower() == 'medium' else 3
)
wines['AcidityRank'] = wines.Acidity.apply(acidityConvertion)
```

Converting Body to numerical ranked {Very light-bodie = 1, Light-bodied = 2, Medium-bodied = 3, Full-bodied = 4, Very full-bodied = 5}

```python
bodyConvertion = lambda x: (
    1 if x.lower() == 'Very light-bodied'.lower() else
    2 if x.lower() == 'Light-bodied'.lower() else
    3 if x.lower() == 'Medium-bodied'.lower() else
    4 if x.lower() == 'Full-bodied'.lower() else 5 
)
wines['BodyRank'] = wines.Body.apply(bodyConvertion)
```

Flat the Harmonize List

```python
wines['Harmonize'] = wines['Harmonize'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
harmonize_df = wines.explode('Harmonize')
harmonize_df.head()
```

Flat the Grapes List

```python
wines['Grapes'] = wines['Grapes'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
grapes_df = wines.explode('Grapes')
grapes_df.head()
```

Flat the Vintage Years List

```python
wines['VintageYears'] = wines['VintageYears'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
years_df = wines.explode('VintageYears')
years_df.head()
```

Merged Wines and Ratings Data Sets

```python
merged_df = pd.merge(wines, wines_ratings, on='WineID')
merged_df.head()
```

The heatmap displays the frequency counts of the top 10 most common food pairings across these wine categories, illustrating which types of wines are most commonly paired with specific dishes.

```python
# Count frequency of each grape
top_h = harmonize_df['Harmonize'].value_counts().head(10).index
filtered_h = harmonize_df[harmonize_df['Harmonize'].isin(top_h)]
crosstab = pd.crosstab(filtered_h['Harmonize'], filtered_h['Type'])

# Plot as a tile using a heatmap)
plt.figure(figsize=(12, 8))
sns.heatmap(crosstab, annot=True, fmt=".0f", cmap='Oranges',
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title("Harmonize Food vs WineType")
plt.xlabel("Wine Type")
plt.ylabel("Harmonize Food")
plt.tight_layout()
plt.show()
```

- Red wine is the dominant pairing choice for most meat dishes, especially Beef (471), Game Meat (311), Lamb (365), and Poultry (387).
- Pork dishes demonstrate versatility in wine pairings, with notable counts for White (114), Sparkling (61), and Rosé (31) wines.

By calculating the average BodyRank for each, we gain insight into how different grape types tend to influence the wine’s body.

```python
# Count frequency of each grape
top_grapes = grapes_df['Grapes'].value_counts().head(10).index

# Filter only rows with those top 10 grapes
filtered = grapes_df[grapes_df['Grapes'].isin(top_grapes)]

# Group By Grapes Type, and calculate the BodyRank mean
mean_body = filtered.groupby('Grapes')['BodyRank'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=mean_body.index, y=mean_body.values, palette='Oranges', hue=mean_body.values)
plt.title('Average Body Ranking by Grape Type')
plt.ylabel('Average Body Rank')
plt.xlabel('Grape Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Touring Nacional has the most complex Body rasing between Full-bodied and Very Full-bodied.
- Merlot and Cabernet Franc turn on the median around 3.6 lining toward Medium-bodied.
