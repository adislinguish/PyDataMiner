<h2>Summary Description</h2>
This dataset aggregates detailed information on wine products and user ratings sourced from the open web. <h3>Data Type Conversion</h3>
How the dataset does not have lot of numeric values, we are going to convert the ranked categorical variables of Acidity and Body to numerical ranking. 
Converting Acidity to numerical ranked {Low = 1, Medium = 2, High = 3}

```python
acidityConvertion = lambda x: (
    1 if x.lower() == 'low' else
    2 if x.lower() == 'medium' else 3
)
wines['AcidityRank'] = wines.Acidity.apply(acidityConvertion)
```

Converting Body to numerical ranked {Very light-bodie = 1, Light-bodied = 2, Medium-bodied = 3, Full-bodied = 4, Very full-bodied = 5}

```python
bodyConvertion = lambda x: (
    1 if x.lower() == 'Very light-bodied'.lower() else
    2 if x.lower() == 'Light-bodied'.lower() else
    3 if x.lower() == 'Medium-bodied'.lower() else
    4 if x.lower() == 'Full-bodied'.lower() else 5 
)
wines['BodyRank'] = wines.Body.apply(bodyConvertion)
```

Flat the Harmonize List

```python
wines['Harmonize'] = wines['Harmonize'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
harmonize_df = wines.explode('Harmonize')
harmonize_df.head()
```

Flat the Grapes List

```python
wines['Grapes'] = wines['Grapes'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
grapes_df = wines.explode('Grapes')
grapes_df.head()
```

Flat the Vintage Years List

```python
wines['VintageYears'] = wines['VintageYears'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
years_df = wines.explode('VintageYears')
years_df.head()
```

Merged Wines and Ratings Data Sets

<h3>Unsupervised Learning</h3>
A clean, binarized, and filtered dataset with selected features optimized for association rule mining Apriori in WEKA.
The preprocessing pipeline began with binarizing and simplified multiple columns such as Grapes, Elaborate, Harmonize and Type.

The following columns were also deleted:
- Acidity, Body: already represented as ranked values like AcidityRank and BodyRank.
- WineID, VintageYears, RegionID, WineryID: identifiers or text fields not useful for generalization.

In WEKA, additional preprocessing was applied:
- Discretization: Continuous attributes (like ABV) were discretized using bins (3 bins with equal frequency).
- Feature Selection: Applied InfoGain + Ranker to retain only the most informative attributes.
- Attribute Removal

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Initialize the binarizer
mlb = MultiLabelBinarizer()

# Fit and transform the Grapes column
grapes_binarized = pd.DataFrame(mlb.fit_transform(wines['Grapes']), 
                                columns=['Grapes_' + col.replace(' ', '') for col in mlb.classes_],
                                index=wines.index)
grapes_binarized = grapes_binarized.replace({1: 'Y', 0: 'N'})   

# Concatenate the binarized columns with the original DataFrame
df_binarized = pd.concat([wines.drop(columns=['Grapes']), grapes_binarized], axis=1)

#Binarizing Harmonize column
# Fit and transform the Harmonize column
harmonize_binarized = pd.DataFrame(mlb.fit_transform(df_binarized['Harmonize']), 
                                columns=['Harmonize_' + col.replace(' ', '') for col in mlb.classes_],
                                index=df_binarized.index)
harmonize_binarized = harmonize_binarized.replace({1: 'Y', 0: 'N'})

# Concatenate the binarized columns with the original DataFrame
df_binarized = pd.concat([df_binarized.drop(columns=['Harmonize']), harmonize_binarized], axis=1)

# Fit and transform the Elaborate column
replacement_map = {
    r'Varietal.*': 'Varietal',
    r'Assemblage.*': 'Assemblage',
   }
# Apply replacements using regex
for pattern, replacement in replacement_map.items():
    df_binarized['Elaborate'] = df_binarized['Elaborate'].str.replace(pattern, replacement, regex=True)
df_binarized['Varietal'] = df_binarized['Elaborate'].apply(lambda x: 'Y' if x == 'Varietal' else 'N')
df_binarized['Assemblage'] = df_binarized['Elaborate'].apply(lambda x: 'Y' if x == 'Assemblage' else 'N')

# Fit and transform the Type column
df_type = df_binarized.copy()
df_type['value'] = 'Y'
df_type = df_type.pivot(index='WineID', columns='Type', values='value').fillna('N').reset_index()
df_binarized = df_binarized.merge(df_type, on='WineID')

del df_binarized['Type']
del df_binarized['Elaborate']
del df_binarized['Acidity']
del df_binarized['Body']
del df_binarized['WineID']
del df_binarized['VintageYears']
del df_binarized['RegionID']
del df_binarized['WineryID']

#Export data to use on Weka
df_binarized.to_csv("wines_transformed_unsupervised.csv", header=True, index=False, quoting=1)
```
