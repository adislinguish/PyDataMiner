<h3>Data Type Conversion</h3>
How the dataset does not have lot of numeric values, we are going to convert the ranked categorical variables of Acidity and Body to numerical ranking. 
Converting Acidity to numerical ranked {Low = 1, Medium = 2, High = 3}

```python
acidityConvertion = lambda x: (
    1 if x.lower() == 'low' else
    2 if x.lower() == 'medium' else 3
)
wines['AcidityRank'] = wines.Acidity.apply(acidityConvertion)
```

Converting Body to numerical ranked {Very light-bodie = 1, Light-bodied = 2, Medium-bodied = 3, Full-bodied = 4, Very full-bodied = 5}

```python
bodyConvertion = lambda x: (
    1 if x.lower() == 'Very light-bodied'.lower() else
    2 if x.lower() == 'Light-bodied'.lower() else
    3 if x.lower() == 'Medium-bodied'.lower() else
    4 if x.lower() == 'Full-bodied'.lower() else 5 
)
wines['BodyRank'] = wines.Body.apply(bodyConvertion)
```

Flat the Harmonize List

```python
wines['Harmonize'] = wines['Harmonize'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
harmonize_df = wines.explode('Harmonize')
harmonize_df.head()
```

Flat the Grapes List

```python
wines['Grapes'] = wines['Grapes'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
grapes_df = wines.explode('Grapes')
grapes_df.head()
```

Flat the Vintage Years List

```python
wines['VintageYears'] = wines['VintageYears'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
years_df = wines.explode('VintageYears')
years_df.head()
```

The preprocessing pipeline began with binarizing and simplified multiple columns such as Grapes, Elaborate, Harmonize and Type.

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Initialize the binarizer
mlb = MultiLabelBinarizer()

# Fit and transform the Grapes column
grapes_binarized = pd.DataFrame(mlb.fit_transform(wines['Grapes']), 
                                columns=['Grapes_' + col.replace(' ', '') for col in mlb.classes_],
                                index=wines.index)
grapes_binarized = grapes_binarized.replace({1: 'Y', 0: 'N'})   

# Concatenate the binarized columns with the original DataFrame
df_binarized = pd.concat([wines.drop(columns=['Grapes']), grapes_binarized], axis=1)

#Binarizing Harmonize column
# Fit and transform the Harmonize column
harmonize_binarized = pd.DataFrame(mlb.fit_transform(df_binarized['Harmonize']), 
                                columns=['Harmonize_' + col.replace(' ', '') for col in mlb.classes_],
                                index=df_binarized.index)
harmonize_binarized = harmonize_binarized.replace({1: 'Y', 0: 'N'})

# Concatenate the binarized columns with the original DataFrame
df_binarized = pd.concat([df_binarized.drop(columns=['Harmonize']), harmonize_binarized], axis=1)

# Fit and transform the Elaborate column
replacement_map = {
    r'Varietal.*': 'Varietal',
    r'Assemblage.*': 'Assemblage',
   }
# Apply replacements using regex
for pattern, replacement in replacement_map.items():
    df_binarized['Elaborate'] = df_binarized['Elaborate'].str.replace(pattern, replacement, regex=True)
df_binarized['Varietal'] = df_binarized['Elaborate'].apply(lambda x: 'Y' if x == 'Varietal' else 'N')
df_binarized['Assemblage'] = df_binarized['Elaborate'].apply(lambda x: 'Y' if x == 'Assemblage' else 'N')

# Fit and transform the Type column
df_type = df_binarized.copy()
df_type['value'] = 'Y'
df_type = df_type.pivot(index='WineID', columns='Type', values='value').fillna('N').reset_index()
df_binarized = df_binarized.merge(df_type, on='WineID')

del df_binarized['Type']
del df_binarized['Elaborate']
del df_binarized['Acidity']
del df_binarized['Body']
del df_binarized['WineID']
del df_binarized['VintageYears']
del df_binarized['RegionID']
del df_binarized['WineryID']

#Export data to use on Weka
df_binarized.to_csv("wines_transformed_unsupervised.csv", header=True, index=False, quoting=1)
```

<h4>What did  not work and Why</h4>

- Dessert and Rosé Classification Weakness
    - Models frequently misclassified Dessert and Rosé:
    - Dessert often collapsed into White or Sparkling.
    - Rosé overlapped with White and light Red wines.
    - Dessert wines share acidity and food pairings with certain Whites.
    - Rosés share grape origins like Pinot Noir with Reds and pairings with Whites.
    - Small sample size makes weaker learnings.

- Clustering Limitations
    - Dessert wines did not form a distinct cluster.
    - Rosé wines spread across clusters with light reds or whites.
    - White clusters looked too similar due to overlapping acidity and seafood pairings.

<h4>Methods That Did Not Work</h4>

- Highly imbalanced classes caused some classifiers to collapse into predicting the majority class, producing deceptively high accuracy but poor per-class recall.
- The preprocessing pipeline removed many attributes, possibly discarding informative features.

<h4>Conclusion</h4>
<p>
    In this data mining project, multiple machine learning techniques were applied to a wine dataset to predict wine type, discover hidden relationships, and understand how grape varieties, alcohol content, body, acidity, and food pairings interact. Across all models tested: J48, JRip, Random Forest, Bagging, Boosting, MLP neural networks, Apriori, and K-Means clustering consistent patterns emerged that highlighted both the strengths and challenges of the dataset.
</p>
<p>
    Supervised learning models performed strongly overall. Ensemble methods such as Bagging and Boosting achieved the highest predictive accuracy 89.05% and 88.56% respectively, showing that combining multiple weak learners leads to more stable and reliable predictions. Random Forest and the MLP neural network also demonstrated excellent performance, although all models struggled with the same classes. Red, White, and Dessert/Port wines were consistently easy to classify due to clear attribute separation, while Dessert and Rosé wines remained difficult for every algorithm. This difficulty originated from overlapping attribute as body, acidity, and food-pairing which caused these wines to be absorbed into the feature space of White or Red wines.
</p>
<p>
    Models like J48 and JRip revealed similar insights. They consistently selected ABV, BodyRank, and key food pairing attributes such as GameMeat, Pork, RichFish as the most influential predictors. These rule based models confirmed that wine type is strongly structured around alcohol level, body, and gastronomic harmony.
</p>  
<p>
    The Apriori association rules exposed additional relationships between grape varieties and food pairings, though many rules had high confidence but low lift, suggesting predictable but not surprising patterns. Clusters of white wines, medium reds, and full-bodied reds emerged clearly, while Dessert and Rosé wines were scattered across clusters resalting their poor classification performance. The clustering also aligned with the food-pairing and grape-variety associations found by Apriori, reinforcing the internal consistency of the dataset.
</p>
